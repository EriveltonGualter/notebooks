## Study of Electromechanical Systems with Energy Regeneration

> *This notes is the result of analysis of the following paper:*<br>
Richter, Hanz, and Dhipak Selvaraj. "Impedance control with energy regeneration in advanced exercise machines." *American Control Conference (ACC)*, 2015. IEEE, 2015.

Driving-point impedance or just impedance is the mechanical propriety that relates the force applied and the motion of the system (velocity and acceleration). The goals is to control the interaction of the system with the environment. It differs of the usual control techniques that has the main goal to achieve a reference point. 

The impedance control has been used in several applications, such as suspension system, humanoid robot interaction, and prosthetic robotics as well. Here, contains the study of impedance control related to Advanced Exercise Machines and in the following study I will present this study for prosthetic application.

### Mathematical Model

For the sum of momentus iqual zero, the system present the following differential equations:

$$ J_t \cdot \dot w_g = \tau_{cable} - k_s\cdot\Theta_g - \Phi_t - \tau_{ind}$$

> $ J_t = n^2\cdot J_m + J_g + J_d $; <br>
$ w_g $ : Drum angular velocity; <br>
$ \tau_{cable} = r \cdot F$ : Applied momentum;<br>
$ k_s$ : Torsional spring constant;<br>
$ \Theta_g$ : Drum angular position; <br>
$ \Phi_t$ : Nonlinear friction function;<br>
$ \tau_{ind}$ : Motor torque

According to fundamental conversion between mechanical and electrical ports, we have: 

$$\tau_{ind} = \alpha\cdot i_m$$
$$e_{ind} = \alpha\cdot\omega_g$$

Also, ponting out the ideal capacitor model:
<figure><img src='images/ideal_charge.png' width="150" alt="" style="float:left;margin: 10 10 10 10px;"/></figure>

$$ C = \frac{q_c}{v_c} $$
$$ i_c = C\cdot\dot v_c = \dot q_c $$

<br>
where, $C$ is the capacitance, $q_c$ is the stored charge and $v_c$ is the voltage capacitor.<br><br>

Therefore, applying Kirchhoff's Voltage Law ($\sum_{}V_i = 0$) in the armature circuit, we have:

$$e_{ind} - i_m\cdot R - v_m = 0$$

Assuming that $ v_m = u\cdot v_c $ and using the fundamental conversion relationship 
$e_{ind} = \alpha\cdot\omega$, we have:

$$e_{ind} - i_m\cdot R - v_m = 0$$
$$\alpha\cdot\omega_g - i_m\cdot R - u\cdot v_c  = 0$$
$$( \alpha\cdot\omega_g - i_m\cdot R - u\cdot v_c )\cdot\alpha  = 0\cdot\alpha$$
$$\alpha^2\cdot\omega_g - R\cdot\alpha\cdot i_m - u\cdot v_c\cdot\alpha = 0$$

Replacying $\tau_{ind} = \alpha\cdot i$ and $v_c = \frac{q_c}{C}$. Then:

$$\tau_{ind} = \frac{\alpha^2}{R}\omega_g - \frac{\alpha\cdot u}{RC}q_c$$

Therefore, the final model is:


$$ \dot w_g = \frac{1}{J_t} \left(r\cdot F(t) - k_s\cdot\Theta_g - \Phi_t - \frac{\alpha^2}{R}\omega_g + \frac{\alpha\cdot u}{RC}q_c\right) $$


---

To find the charge dynamic model of capacitor, it is necessary to go back on the armature circuit:

$$e_{ind} - i_m\cdot R - v_m = 0$$

Because we area assuming that the converter is 100% efficient, the relantionship between the motor and capacitor current is: 

$$ u = \frac{i_c}{i_m} $$


Therefore:

$$\alpha\cdot\omega_g - \frac{i_c}{u} \cdot R - u \cdot v_c = 0$$

Multiplying by $u$ and replacing $v_c$ per $\frac{q_c}{C}$ and $i_c$ per $\dot q_c$:

$$\alpha\cdot\omega_g\cdot u - \dot q_c \cdot R - u^2 \cdot \frac{q_c}{C} = 0$$

Therefore the Dynamic Model of Capacitor is:

$$ \dot q_c = \frac{\alpha\cdot u}{R}\omega_g - \frac{u^2}{RC}q_c$$


### Conclusion
$$ \dot\Theta_g = \omega_g$$
$$ \dot w_g = \frac{1}{J_t} \left(r\cdot F(t) - k_s\cdot\Theta_g - \Phi_t - \frac{\alpha^2}{R}\omega_g + \frac{\alpha\cdot u}{RC}q_c\right) $$
$$ \dot q_c = \frac{\alpha\cdot u}{R}\omega_g - \frac{u^2}{RC}q_c$$


> The paper has a mistake definition for "u". It says that the $ v_c = u\cdot v_m $; however, the right form is: $ v_m = u\cdot v_c $
